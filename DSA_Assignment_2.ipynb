### Calculators and their Operations

##### 1. Calculator

```python
# Function to perform basic arithmetic operations
def calculator(a, b, operation):
    # Check the operation and perform accordingly
    if operation == '+':
        return a + b  # Addition
    elif operation == '-':
        return a - b  # Subtraction
    elif operation == '*':
        return a * b  # Multiplication
    elif operation == '/':
        # Handle division by zero
        return a / b if b != 0 else "Cannot divide by zero"
    # Square operation (only needs one number)
    elif operation == 'square':
        return a ** 2  # Square of number
    else:
        return "Invalid operation"

print(calculator(10, 5, '+'))
print(calculator(10, 5, '/'))
print(calculator(84 ,0, 'square'))
print(calculator(16, 2, 'log'))
```

##### 2. Discount Calculator

```python
# Function to calculate final price after discount
def discount(price, discount_percent):

    # Calculate discount amount
    discount_amount = (price * discount_percent) / 100

    # Subtract discount from original price
    final_price = price - discount_amount
    return final_price

#Taking user input
price = float(input("Enter original price: "))
discount_percent = float(input("Enter discount percentage: "))

# Call the function
final_price = discount(price, discount_percent)

# Display results
print("Original Price:", price)
print("Discount:", discount_percent, "%")
print("Final Price after discount:", final_price)
```

##### 3. Grade Evaluator

```python
# Function to assign grade based on marks
def grade(marks):
    if marks >= 90:
        return "A"  # Excellent
    elif marks >= 75:
        return "B"  # Good
    elif marks >= 60:
        return "C"  # Average
    elif marks >= 40:
        return "D"  # Pass
    else:
        return "Fail"  # Fail

#Taking user input
marks = float(input("Enter your marks: "))
# Calling function
result = grade(marks)
# Display result
print("Your Grade is:", result)
```

##### 4. Even / Odd Checker

```python
# Function to check whether a number is even or odd
def even_odd(num):

    # Reject negative numbers
    if num < 0:
        return "Negative numbers are not allowed"

    # Check if number is divisible by 2
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"


#Taking user input
num = int(input("Enter a number: "))
# Calling function
result = even_odd(num)
# Display result
print("Result:", result)
```

### Array (List) Operations

#### 1. Swapping Elements

```python
# Function to swap two elements in a list
def swap(arr, i, j):
    # Swap using tuple unpacking
    arr[i], arr[j] = arr[j], arr[i]
    return arr

# Example
arr = [1, 2, 3, 4]
print(swap(arr, 0, 2))
```

#### 2. Splitting Elements

```python
# Function to split array into two halves
def split_array(arr):
    # Find middle index
    mid = len(arr) // 2

    # Return two halves
    return arr[:mid], arr[mid:]

# Example
arr = [1, 2, 3, 4, 5]
print(split_array(arr))  # Output: ([1, 2], [3, 4, 5])
```

#### 3. Inserting Elements

```python
# Function to insert an element at a given index
def insert_element(arr, index, value):
    # Insert value at specified position
    arr.insert(index, value)
    return arr

# Example
arr = [1, 2, 4]
print(insert_element(arr, 2, 3))
```

#### 4. Deleting Elements

```python
# Function to delete a specific value from list
def delete_element(arr, value):
    # Check if value exists in list
    if value in arr:
        arr.remove(value)  # Remove first occurrence
    return arr

# Example
arr = [1, 2, 3, 4]
print(delete_element(arr, 3))  # Output: [1, 2, 4]
```
