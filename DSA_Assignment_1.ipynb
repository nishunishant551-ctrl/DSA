# Assignment : 1
You sweet and innocent folks are enrolled as BSC-AI students for a 3 year degree program. You are supposed to study a total of 10 subjects in each particular year. Create a 3D array to represent/store student marks in the form of Data where Shape: (Year, Student, Subject)
Example -> shape: (3 years, 16 students, 10 subjects)
## Tasks:
Initialize using np.arange() and reshape().  (optional)
Compute:
* Year-wise average
* Student-wise total
* Subject-wise highest score
* Replace all marks below 40 with 40 (Updating operation).
* Sort marks subject-wise.


# Step 1: Create 3D Array (Year, Student, Subject)

```python
import numpy as np
#Shape: (3 years, 16 students, 10 subjects)
data = np.arange(3 * 16 * 10).reshape(3, 16, 10)

print("3D Data:", data)
print("Shape:", data.shape)
```

# Step 2: Year-wise Average
Average marks per year (across all students & subjects)

```python
#Year-wise Average
year_avg = data.mean(axis=(1, 2))
print("Year-wise Average:", year_avg)
```

#Step 3: Student-wise Total
Total marks of each student (across all subjects, per year)

```python
#Student-wise Total
student_total = data.sum(axis=2)
print("Student-wise Total:", student_total)
```

#Step 4: Subject-wise Highest Score
Highest marks in each subject across all students & years

```python
# Subject-wise Highest Score
subject_max = data.max(axis=(0, 1))
print("Subject-wise Highest Score:", subject_max)
```

#Step 5: Replace Marks Below 40
Any value < 40 becomes 40

```python
#Replace Marks Below 40 with 40
data[data < 40] = 40
print("Updated Data:", data)
```

#Step 6: Sort Marks Subject-wise
Sort marks along subjects (last axis)

```python
#Sort Marks Subject-wise
sorted_data = np.sort(data, axis=1)
print("Subject-wise Sorted Data:\n", sorted_data)
```
