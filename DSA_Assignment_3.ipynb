### Question:   Implement a Python program to solve quadratic equations with condition checks, write and compare Selection, Insertion, and Bubble Sort algorithms, implement Quick Sort using divide and conquer.

### The Quadratic Equation

```python
#Function to solve the quadratic equation

def quadratic(a, b, c):

  #First, we check if "a" is zero or not. It is an essential condition.
  if a==0:
    return "Not a quadratic equation"

  # Discriminant
  d = b**2 - 4*a*c

  if d>0:
    root1 = (-b + d**0.5) / (2*a)
    root2 = (-b - d**0.5) / (2*a)
    return f"Real and distinct roots: {root1}, {root2}"

  elif d == 0:
        root = -b / (2*a)
        return f"Real and equal roots: {root}"

  else:
        # Complex roots
        real = -b / (2*a)
        imag = (-d)**0.5 / (2*a)
        return f"Complex roots: {real} ± {imag}i"


# Input
a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))

print(quadratic(a, b, c))
```

### The Sorting Algorithms

#### 1. Selection Sort

```python
def selection_sort(arr):
    for i in range(len(arr)):
        min_index = i

        for j in range(i+1, len(arr)):
            if arr[j] < arr[min_index]:
                min_index = j

        arr[i], arr[min_index] = arr[min_index], arr[i]

    return arr
```

#### 2. Insertion Sort

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
    return arr
```

#### 3. Bubble Sort

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
```

#### 4. Quick Sort

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[0]
    left = []
    right = []

    for x in arr[1:]:
        if x <= pivot:
          left.append(x)
        else:
            right.append(x)

    return quick_sort(left) + [pivot] + quick_sort(right)
```

```python
# Testing Sorting
arr = [5, 3, 8, 4, 2]

print("\nOriginal Array:", arr)
print("Selection Sort:", selection_sort(arr.copy()))
print("Insertion Sort:", insertion_sort(arr.copy()))
print("Bubble Sort:", bubble_sort(arr.copy()))
print("Quick Sort:", quick_sort(arr.copy()))
```
